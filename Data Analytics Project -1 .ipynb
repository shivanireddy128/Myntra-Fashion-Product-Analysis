## Data Pre-Processing

__Feature Engineering__

```python
import ast
def parse_attributes(x):
    if isinstance(x,str):
        try:
            return ast.literal_eval(x)
        except:
            return {}
    else:
        return {}
#Apply function to the column
df["p_attributes"] = df["p_attributes"].apply(parse_attributes)
#Extract columns from the dictionary
df["Occasion"] = df["p_attributes"].apply(lambda x: x.get("Occasion","Unknown"))
df["Top Fabric"] = df["p_attributes"].apply(lambda x: x.get("Top Fabric","Unknown"))
```

## Insights

__1. Products priced between ₹1000 and ₹2000 tend to have higher average ratings compared to lower and higher price ranges,
indicating that customers rate mid-priced products more positively__

__2. Brands in the mid-price range generally maintain higher and more consistent ratings compared to premium-priced brands,
suggesting stronger perceived value for money__

__3. Certain colours such as Black, Navy Blue, and Red appear more frequently in highly rated products,
indicating that these colours are commonly offered among well-rated items__

__4. Brands with consistently high average ratings (above 4.2) demonstrate stronger customer trust, as reflected through rating patterns__
